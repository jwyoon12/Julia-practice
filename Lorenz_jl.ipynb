```julia
using Plots
using DifferentialEquations
```

```julia
function lorenz!(du, u, p, t)
    x, y, z = u
    σ, ρ, β = p
    du[1] = σ*(y-x)
    du[2] = x*(ρ-z)-y
    du[3] = x*y-β*z
    
end
```

```julia
u0 = [1.0,0.0,0.0]
tspan = (0.0,100.0)
p = [10.0, 28.0, 8/3]

prob = ODEProblem(lorenz!, u0, tspan, p)
sol = solve(prob)
```

```julia
plot1 = plot!(sol, 
                vars=(1,2,3),
                xlabel = "x",
                ylabel = "y",
                zlabel = "z",
                label = "x against y against z")
```
